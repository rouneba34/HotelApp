```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
#Télécharger le fichier csv à partir du lien github
url='https://raw.githubusercontent.com/shivanpillay20/dataset/refs/heads/main/autompg.csv'
dataset=pd.read_csv(url, encoding='latin1')
print(dataset)
```

```python
#Valeurs manquantes
missing_values = dataset.isnull().sum()
print(missing_values)
```

```python
dataset['accélération'] = pd.to_numeric(dataset['accélération'], errors='coerce')
dataset['année du modèle'] = pd.to_numeric(dataset['année du modèle'], errors='coerce')
dataset['Numéro du véhicule'] = pd.to_numeric(dataset['Numéro du véhicule'], errors='coerce')
```

```python
#Supprimer les valeurs abérantes
dataset=dataset.dropna(subset=['accélération'])

# Boxplot for 'accélération'
plt.figure(figsize=(6, 4))
sns.boxplot(x=dataset['accélération'])
plt.title('Boxplot of Accélération')
plt.xlabel('accélération')
plt.show()

```

```python
#Supprimer les valeurs abérantes
dataset=dataset.dropna(subset=['année du modèle'])

# Boxplot for 'année du modèle'
plt.figure(figsize=(6, 4))
sns.boxplot(x=dataset['année du modèle'])
plt.title('Boxplot of Année du modèle')
plt.xlabel('année du modèle')
plt.show()
```

```python
#Supprimer les valeurs abérantes
dataset=dataset.dropna(subset=['Numéro du véhicule'])

# Boxplot for 'numéro du véhicule'
plt.figure(figsize=(6, 4))
sns.boxplot(x=dataset['Numéro du véhicule'])
plt.title('Boxplot of Numéro du véhicule')
plt.xlabel('Numéro du véhicule')
plt.show()
```
